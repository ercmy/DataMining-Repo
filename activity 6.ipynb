```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import warnings
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('dengue.csv')
```

```python
dengue
```

```python
dengue.isnull().sum()
```

```python
dengue.dtypes
```

```python
dengue['Month'] = dengue['Month'].astype('category')
dengue['Year'] = dengue['Year'].astype('category')
dengue['Region'] = dengue['Region'].astype('category')
```

```python
dengue.dtypes
```

<h1>Insight 1</h1>

```python
visual1 = dengue.groupby(['Month', 'Year'])['Dengue_Cases'].sum().reset_index()

x = visual1['Year']
y = visual1['Dengue_Cases']

plt.bar(x, y, color='brown')
plt.gcf().set_size_inches(12, 8)
plt.xlabel("Year")
plt.ylabel("Dengue Cases")
plt.title("Dengue Cases Per Year")
plt.show()
```

<h3>Insight 1.1</h3>
<h4>The year with a highest dengue cases is year 2019</h4>

<h1>Insight 18</h1>

```python
trend_data = dengue.groupby(["Year", "Month"])[["Dengue_Cases"]].sum().reset_index()

plt.figure(figsize=(12, 6))

sns.lineplot(data=trend_data, x="Month", y="Dengue_Cases", hue="Year", marker="o", linewidth=2)

plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Dengue Cases")
plt.title("Dengue Cases Trend Over Time")
plt.show()
```

<h3>Insight 18.1</h3>
<h4>In 2019, dengue cases were significantly higher compared to other years, indicating a major outbreak. </h4>

<h1>Insight 19</h1>

```python
heatmap_data = dengue.pivot_table(index="Year", columns="Month", values="Dengue_Cases", aggfunc="sum")

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap="Reds", annot=True, fmt=".0f")

plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Dengue Cases Heatmap (Seasonal Trend)")
plt.show()
```

<h3>Insight 19.1</h3>
<h4>In 2019 of September(89,642 cases) and August(85,038 cases) showing the highest recorded cases</h4>

<h1>Insight 20</h1>

```python
data = dengue.pivot_table(index="Year", columns="Month", values="Dengue_Deaths", aggfunc="sum")

plt.figure(figsize=(12, 6))
sns.heatmap(data, cmap="YlGnBu", annot=True, fmt=".0f")

plt.xlabel("Month")
plt.ylabel("Year")
plt.title("Dengue Deaths Heatmap (Seasonal Trend)")
plt.show()
```

<h3>Insight 20.1</h3>
<h4>In 2016 of October(3954 deaths) and September(3418 deaths) showing the highest recorded deaths</h4>
